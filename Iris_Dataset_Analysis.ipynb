Data Visvalization

```python
#data visualization

#Scatter Plot (Relationship between features)
sns.scatterplot(x='Sepal.Length', y='Sepal.Width', hue='Species', data=df) # relation between Lenght and Width
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
#Histogram(tells how data is spread, skeweness/ symmetry )
df.hist(figsize=(10,8)) # size of figure is adjusted inside hist function
plt.suptitle("Feature Distributions")
plt.show()
```

```python
#Box plot 
plt.figure(figsize=(10,6))
sns.boxplot(data=df) # detects outliers 
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
sns.pairplot(df, hue='Species')
plt.show()
# this shows all features relation at once 
```
